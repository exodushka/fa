Создайте программу, которая выполняет следующие действия:</br>а) Возвращает словарь, отсортированный по значениям в порядке возрастания и убывания. _Пример_: `{1:2, 3:4, 4:3, 2:1, 0:0}` → `{0:0, 2:1, 1:2, 4:3, 3:4}`.</br>б) Объединяет два словаря, при этом суммирует значения для одинаковых ключей. _Пример_: `{1:2, 3:4}` и `{3:5, 6:7}` → `{1:2, 3:9, 6:7}`.</br>в) Проверяет наличие определенных ключей в словаре. _Пример_: `{1:2, 3:4}` и `[1, 3]` → `True`.

```python
def a(d: dict):
    copy_dict = dict()
    list_from_dict = []
    for i in d:
        index, value = i, d[i]
        list_from_dict.append([index, value])
    
    for i in range(len(list_from_dict)):
        for j in range(i, len(list_from_dict)):
            if list_from_dict[i][0] > list_from_dict[j][0]:
                temp = list_from_dict[i]
                list_from_dict[i] = list_from_dict[j]
                list_from_dict[j] = temp
    for i1, i2 in list_from_dict:
        copy_dict[i1] = i2
    return copy_dict

def b(d1: dict, d2: dict):
    for index in d2:
        if index in d1:
            d1[index] += d2[index]
        else:
            d1[index] = d2[index]
    return d1

def c(d: dict, value: list):
    for i in value:
        if i not in d:
            return False
    return True
test_dict_a = {
    2: 1,
    4: 5,
    1: 6,
    3: 7
}
q = (a(test_dict_a))
print("пункт а")
for i in q:
    print(i, q[i])



test_dict_b1 = {
    1:2,
    3:4,
}
test_dict_b2 = {
    3:5,
    6:7
}
print()
print("пунтк б")
q = (b(test_dict_b1, test_dict_b2))
for i in q:
    print(i, q[i])



test_dict_c = {
    1:2,
    3:4
}
test_list = [1,3]
print()
print("пунтк в")
print(c(test_dict_c, test_list))
```
