### pickle

2.1 На основе файла `contributors_sample.json` создайте словарь следующего вида: 
```
{
    должность: [список username людей, занимавших эту должность]
}
```

````python
with open('data/contributors_sample.json', 'r', encoding='utf-8') as f:
    contributors = json.load(f)

job_people = {}

for user in contributors:
    for job in user['jobs']:
        if job not in job_people:
            job_people[job] = []
        job_people[job].append(user['username'])

print(f"Всего уникальных должностей: {len(job_people)}")
print(f"\nПримеры должностей с количеством людей:")
for job, usernames in list(job_people.items())[:10]:
    print(f"  {job}: {len(usernames)} чел.")
````

2.2 Сохраните результаты в файл `job_people.pickle` и в файл `job_people.json` с использованием форматов pickle и JSON соответственно. Сравните объемы получившихся файлов. При сохранении в JSON укажите аргумент `indent`.

````python
with open('data/contributors_sample.json', 'r', encoding='utf-8') as f:
    contributors = json.load(f)

job_people = {}
for user in contributors:
    for job in user['jobs']:
        if job not in job_people:
            job_people[job] = []
        job_people[job].append(user['username'])

with open('job_people.pickle', 'wb') as f:
    pickle.dump(job_people, f)

with open('job_people.json', 'w', encoding='utf-8') as f:
    json.dump(job_people, f, ensure_ascii=False, indent=2)

pickle_size = os.path.getsize('job_people.pickle')
json_size = os.path.getsize('job_people.json')

print(f"Размер job_people.pickle: {pickle_size} байт ({pickle_size/1024:.2f} KB)")
print(f"Размер job_people.json: {json_size} байт ({json_size/1024:.2f} KB)")
print(f"\nJSON больше в {json_size/pickle_size:.2f} раз")
print(f"Разница: {json_size - pickle_size} байт")
````

### XML

3.1 По данным файла `steps_sample.xml` сформируйте словарь с шагами по каждому рецепту вида `{id_рецепта: ["шаг1", "шаг2"]}`. Сохраните этот словарь в файл `steps_sample.json`

````python
tree = ET.parse('data/steps_sample.xml')
root = tree.getroot()

recipe_steps = {}

for recipe in root.findall('.//recipe'):
    recipe_id = recipe.get('id')
    steps = []

    for step in recipe.findall('.//step'):
        steps.append(step.text)

    recipe_steps[recipe_id] = steps

with open('steps_sample.json', 'w', encoding='utf-8') as f:
    json.dump(recipe_steps, f, ensure_ascii=False, indent=2)

print(f"Всего рецептов: {len(recipe_steps)}")
print(f"Данные сохранены в steps_sample.json")
````

3.2 По данным файла `steps_sample.xml` сформируйте словарь следующего вида: `кол-во_шагов_в_рецепте: [список_id_рецептов]`

````python
tree = ET.parse('data/steps_sample.xml')
root = tree.getroot()

# Создание словаря {кол-во_шагов: [id_рецептов]}
steps_count_dict = {}

for recipe in root.findall('.//recipe'):
    recipe_id = recipe.get('id')
    steps = recipe.findall('.//step')
    steps_count = len(steps)

    if steps_count not in steps_count_dict:
        steps_count_dict[steps_count] = []
    steps_count_dict[steps_count].append(recipe_id)

# Вывод результатов
print("Распределение рецептов по количеству шагов:")
for count in sorted(steps_count_dict.keys()):
    print(f"  {count} шагов: {len(steps_count_dict[count])} рецептов")
    print(f"    IDs: {steps_count_dict[count][:5]}{'...' if len(steps_count_dict[count]) > 5 else ''}")
````

3.3 Получите список рецептов, в этапах выполнения которых есть информация о времени (часы или минуты). Для отбора подходящих рецептов обратите внимание на атрибуты соответствующих тэгов.

````python
tree = ET.parse('data/steps_sample.xml')
root = tree.getroot()

# Поиск рецептов с информацией о времени
recipes_with_time = []

for recipe in root.findall('.//recipe'):
    recipe_id = recipe.get('id')
    has_time = False

    # Проверка атрибутов шагов на наличие времени
    for step in recipe.findall('.//step'):
        # Проверка атрибутов (например, time, duration, hours, minutes)
        if any(attr in step.attrib for attr in ['time', 'duration', 'hours', 'minutes']):
            has_time = True
            break
        # Проверка текста шага на упоминание времени
        if step.text and any(word in step.text.lower()
                            for word in ['час', 'минут', 'hour', 'minute', 'min', 'hr']):
            has_time = True
            break

    if has_time:
        recipes_with_time.append(recipe_id)

print(f"Рецептов с информацией о времени: {len(recipes_with_time)}")
print(f"IDs: {recipes_with_time[:10]}{'...' if len(recipes_with_time) > 10 else ''}")
````

3.4 Загрузите данные из файла `recipes_sample.csv` (__ЛР2__) в таблицу `recipes`. Для строк, которые содержат пропуски в столбце `n_steps`, заполните этот столбец на основе файла  `steps_sample.xml`. Строки, в которых столбец `n_steps` заполнен, оставьте без изменений.

````python
import pandas as pd
import xml.etree.ElementTree as ET

# Чтение CSV файла без заголовков с указанием имен столбцов
column_names = ['name', 'id', 'prep_time', 'cook_time', 'date', 'n_steps', 'description', 'rating']
recipes = pd.read_csv('/Users/iliaegorov/Projects/fa/datat/02_pandas/data/recipes_sample.csv',
                      names=column_names,
                      header=None,
                      encoding='utf-8')

# Преобразование ID в строку для корректного сопоставления
recipes['id'] = recipes['id'].astype(str)

# ВАЖНО: Преобразуем n_steps к числовому типу (float, чтобы сохранить NaN)
recipes['n_steps'] = pd.to_numeric(recipes['n_steps'], errors='coerce')

print(f"Всего записей: {len(recipes)}")
print(f"Пропусков в n_steps до заполнения: {recipes['n_steps'].isna().sum()}")
print(f"Тип данных n_steps: {recipes['n_steps'].dtype}")

# Парсинг XML файла с шагами рецептов
tree = ET.parse('data/steps_sample.xml')
root = tree.getroot()

# Создание словаря {id_рецепта: количество_шагов}
steps_count = {}
for recipe in root.findall('recipe'):
    recipe_id_elem = recipe.find('id')
    if recipe_id_elem is not None and recipe_id_elem.text:
        recipe_id = str(recipe_id_elem.text).strip()
        steps = recipe.findall('.//step')
        steps_count[recipe_id] = len(steps)

print(f"Рецептов в XML: {len(steps_count)}")

# Заполнение пропусков в столбце n_steps
filled_count = 0
for idx in recipes[recipes['n_steps'].isna()].index:
    recipe_id = recipes.loc[idx, 'id']
    if recipe_id in steps_count:
        recipes.loc[idx, 'n_steps'] = steps_count[recipe_id]
        filled_count += 1

print(f"Заполнено пропусков: {filled_count}")
print(f"Пропусков в n_steps после заполнения: {recipes['n_steps'].isna().sum()}")
````
