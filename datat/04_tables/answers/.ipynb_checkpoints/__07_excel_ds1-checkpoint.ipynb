## Разминка

```python
import xlwings as xw
from xlwings.constants import AutoFillType
import numpy as np
import pandas as pd
```

1. На листе "Рецептура" файла `себестоимостьА_в1.xlsx` для области "Пшеничный хлеб" рассчитать себестоимость всех видов продукции.

```python
wb = xw.Book('себестоимостьА_в1.xlsx')
recepies = wb.sheets['Рецептура']
```

```python
consumption = recepies.range('G7:O10').options(np.array).value
consumption = np.nan_to_num(consumption)
unit_prices = recepies.range('G14:O14').value
cost = np.sum(consumption * unit_prices, axis=1)
```

2. Результаты расчетов 1.1 сохранить в отдельном столбце области "Пшеничный хлеб"

```python
recepies.range('T6').value = 'Себестоимость'
recepies.range('T7').options(transpose=True).value = cost
```

3. Приблизить форматирование столбца, добавленного в задаче 2 к оформлению всей области.

```python
xw.Range('T4:T6').api.Merge()
```

```python
recepies.range('B3:T3').api.Merge()
recepies.range('B3').value = 'Пшеничный хлеб'

xw.Range('T4:T6').api.Merge()
recepies.range('T4').color = recepies.range('S4').color
recepies.range('T4').autofit()
recepies.range('T4').api.Font.ColorIndex = 2
recepies.range('T4').api.VerticalAlignment = xw.constants.VAlign.xlVAlignCenter
recepies.range('T7:T10').number_format = '0,00'
recepies.range('T7:T10').color = recepies.range('S7').color
```

4. Выполнить 3 с помощью "протягиваемых" формул.

```python
recepies.range('U7').formula = '=SUMPRODUCT(G7:O7, $G$14:$O$14)'
xw.Range('U7').api.AutoFill(xw.Range("U7:U10").api, xw.constants.AutoFillType.xlFillDefault)
```

## Лабораторная работа 7

```python
import xlwings as xw
from xlwings.constants import AutoFillType
import numpy as np
import pandas as pd
```

1. Загрузите данные из файлов `reviews_short.csv` и `recipes_short_5.csv` в виде `pd.DataFrame`. Обратите внимание на корректное считывание столбца(ов) с индексами. Оставьте в таблице с рецептами следующие столбцы: `id`, `name`, `minutes`, `submitted`, `description`, `n_ingredients`

```python
recipes = pd.read_csv(r"E:\Downloads\TOBD_2021_datasets\food_com\data\recipes_short.csv", 
                      usecols=['id', 'name', 'minutes', 'submitted', 'description', 'n_ingredients'])
reviews = pd.read_csv(r"E:\Downloads\TOBD_2021_datasets\food_com\data\reviews_short.csv", index_col=0)
```

2. Случайным образом выберите 5% строк из каждой таблицы и сохраните две таблицы на разные листы в один файл `recipes.xlsx`. Дайте листам названия "Рецепты" и "Отзывы", соответствующие содержанию таблиц. 

```python
with pd.ExcelWriter('E:/downloads/recipes.xlsx', engine='xlsxwriter') as writer:
    recipes.sample(frac=.05).to_excel(writer, sheet_name='Рецепты')
    reviews.sample(frac=.05).to_excel(writer, sheet_name='Отзывы')
```

3. Используя `xlwings`, добавьте на лист `Рецепты` столбец `seconds_assign`, показывающий время выполнения рецепта в секундах. Выполните задание при помощи присваивания массива значений диапазону ячеек.

```python
wb = xw.Book('E:/downloads/recipes.xlsx')
recipes = wb.sheets['Рецепты']
reviews = wb.sheets['Отзывы']

n_rows_recipes = recipes.range('A1').current_region.last_cell.row
n_rows_reviews = reviews.range('A1').current_region.last_cell.row

recipes.range("I1").value = 'seconds'
values = recipes.range('D2').expand('down').options(np.array).value * 60
recipes.range("I2").value = values.reshape(-1, 1)
```

4. Используя `xlwings`, добавьте на лист `Рецепты` столбец `seconds_formula`, показывающий время выполнения рецепта в секундах. Выполните задание при формул Excel.

```python
recipes.range("H2").formula = '=D2 * 60'
range_to_fill = recipes.range(f"H2:H{n_rows_recipes}")
recipes.range("H1").value = 'seconds'
recipes.range("H2").api.AutoFill(range_to_fill.api, AutoFillType.xlFillDefault)
```

5. Добавьте на лист `Рецепты`  столбец `n_reviews`, содержащий кол-во отзывов для этого рецепта

```python
recipes.range("J1").value = 'n_reviews'
recipes.range("J2").formula = f'=COUNTIF(Отзывы!$C$2:$C${n_rows_reviews},Рецепты!C2)'
range_to_fill = recipes.range(f"J2:J{n_rows_recipes}")
recipes.range("J2").api.AutoFill(range_to_fill.api, AutoFillType.xlFillDefault)
```

6. Раскрасьте ячейки столбца `minutes` в соответствии со следующим правилом: если рецепт выполняется быстрее 5 минут, то цвет - зеленый; от 5 до 10 минут - жёлтый; и больше 10 - красный.

```python
for x in recipes.range('D2').expand('down'):
    if x.value < 5:
        x.color = (0, 160, 0)
    elif 5 <= x.value <= 10:
        x.color = (236, 245, 66)
    else:
        x.color = (160, 0, 0)
        
```

7. Сделайте названия всех добавленных столбцов полужирными и выровняйте по центру ячейки.

```python
recipes.range("I1").value = 'n_reviews'
recipes.range("I1").api.Font.Bold = True
recipes.range("I1").api.HorizontalAlignment  = xw.constants.HAlign.xlHAlignCenter
```

8. Напишите функцию `validate()`, которая проверяет соответствие всех строк из листа `interactions` следующим правилам:
    * Рейтинг - это число от 0 до 5 включительно
    * Соответствующий рецепт имеется на листе рецептов
    
В случае несоответствия этим правилам, выделите строку красным цветом

```python
def validate():
    colors = {True: (0, 100, 0), False: (100, 0, 0)}
    recipes_id = recipes.range(f"C2:C{n_rows_recipes}").options(np.array, numbers=int).value

    reviews_rec = reviews.range(f"C2:C{n_rows_reviews}").options(np.array, numbers=int).value
    reviews_rat = reviews.range(f"E2:E{n_rows_reviews}").options(np.array, numbers=int).value
    mask = np.isin(reviews_rec, recipes_id) & (reviews_rat <= 5) & (reviews_rat >= 0)
    
    range_to_color = reviews.range('A2').expand()
    for idx, color_flag in enumerate(mask):
        range_to_color.rows[idx].color = colors[color_flag]
```

```python
validate()
```
