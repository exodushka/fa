## Разминка

1. Вывести на экран данные из словаря `obj` построчно в виде `k = v`, задав формат таким образом, чтобы знак равенства оказался на одной и той же позиции во всех строках. Строковые литералы обернуть в кавычки.

````python
obj = {  
         'home_page': 'https://github.com/pypa/sampleproject',
         'keywords': 'sample setuptools development',
         'license': 'MIT',
         'maintainer': None,
         'maintainer_email': None
      }
````

````python
max_len_key = max(len(key) for key in obj)
for k, v in obj.items():
    print('{k:<{max_len}} = "{v}"'.format(k=k, v=v, max_len=max_len_key))
````

2. Дана строка 'aaa--bbb==ccc__ddd'. Написать регулярное выражение для разбивки строки на список ['aaa','bbb','ccc','ddd'].

````python
s = 'aaa--bbb==ccc__ddd'
p = re.compile(r'[a-z]{3}')
p.findall(s)
````

    ```
    |    id     | n_in  |

1.3 Напишите функцию `show_info`, которая для рецепта по его `id` создает описание (в виде одной строки) и выводит ее на экран:

```
"Название"

1. Шаг 1.
2. Шаг 2.
----------
#тэг1 #тэг2
```

````python
def show_info(id_):
    s = f'"{names[int(id_)]}"\n\n'
    s += '\n'.join(f'{idx}. {step.capitalize()}' for idx, step in enumerate(steps[id_], 1))
    s += '\n'
    s += '-'*10
    s += '\n'
    s += ' '.join(f'#{tag}' for tag in tags[int(id_)])
    return s
````

````python
print(show_info('306168'))
````

2.2 Найдите все смайлики в отзывах. Смайлик состоит из трех частей: глаза (символ `=` или `:`), нос (символ `-`), губы (символ `)` или `(`). Смайлик может иметь вид "глаза-нос-губы" или "губы-нос-глаза". Нос может отсутствовать.

````python
patt = re.compile(r'([:=]\-?[\)\(]|[\)\(]\-?[:=])')
s = set()
for d in descriptions:
    r = patt.findall(d) 
    if r:
        s |= set(r)
s
````

4. Напишите функцию, которая для заданного предложения при помощи `pos_tag` выводит информацию о частях речи слов, входящих в предложение в следующем виде:
```
PRP   VBD   DT      NNS     CC   VBD      NNS        RB   
 I  omitted the raspberries and added strawberries instead
``` 


````python
# sent = 'I omitted the raspberries and added strawberries instead'
sent = descriptions[123]
words = word_tokenize(sent)
word_pos = pos_tag(words)
words, pos = zip(*word_pos)
````

````python
spaces = [max(len(word), len(pos)) for word, pos in zip(words, pos)]
print(' '.join('{pos:^{space}}'.format(pos=pos, space=space) for pos, space in zip(pos, spaces)))
print(' '.join('{word:^{space}}'.format(word=word, space=space) for word, space in zip(words, spaces)))
````
