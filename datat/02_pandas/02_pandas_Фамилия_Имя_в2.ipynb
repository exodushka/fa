### Группировки таблиц `pd.DataFrame`

4.1 Посчитайте количество рецептов, представленных каждым из участников (`contributor_id`). Какой участник добавил максимальное кол-во рецептов?

```python
contributor_counts = recipes.groupby('contributor_id').size().reset_index(name='count')
max_contributor = contributor_counts.loc[contributor_counts['count'].idxmax()]
print('Max contributor ID and count')
print(max_contributor)
```

4.2 Посчитайте средний рейтинг к каждому из рецептов. Для скольких рецептов отсутствуют отзывы? Обратите внимание, что отзыв с нулевым рейтингом или не заполненным текстовым описанием не считается отсутствующим.

```python
avg_ratings = reviews.groupby('recipe_id')['rating'].mean().reset_index(name='avg_rating')
recipes_with_reviews_count = len(avg_ratings)
recipes_without_reviews_count = len(recipes) - recipes_with_reviews_count
print('Recipes without reviews')
print(recipes_without_reviews_count)
```

### Объединение таблиц `pd.DataFrame`

5.1 При помощи объединения таблиц, создайте `DataFrame`, состоящий из четырех столбцов: `id`, `name`, `user_id`, `rating`. Рецепты, на которые не оставлен ни один отзыв, должны отсутствовать в полученной таблице. Подтвердите правильность работы вашего кода, выбрав рецепт, не имеющий отзывов, и попытавшись найти строку, соответствующую этому рецепту, в полученном `DataFrame`.

```python
recipes_with_ratings = recipes[['id', 'name']].merge(
    reviews[['recipe_id', 'user_id', 'rating']],
    left_on='id',
    right_on='recipe_id',
    how='inner'
)[['id', 'name', 'user_id', 'rating']]

print('Recipes with ratings shape')
print(recipes_with_ratings.shape)

recipe_ids_with_reviews = set(reviews['recipe_id'])
all_recipe_ids = set(recipes['id'])
recipe_without_review = list(all_recipe_ids - recipe_ids_with_reviews)[0]
is_present = recipe_without_review in recipes_with_ratings['id'].values
print('Recipe without review present in merged table')
print(is_present)
```

5.2 При помощи объединения таблиц и группировок, создайте `DataFrame`, состоящий из трех столбцов: `recipe_id`, `name`, `review_count`, где столбец `review_count` содержит кол-во отзывов, оставленных на рецепт `recipe_id`. У рецептов, на которые не оставлен ни один отзыв, в столбце `review_count` должен быть указан 0. Подтвердите правильность работы вашего кода, выбрав рецепт, не имеющий отзывов, и найдя строку, соответствующую этому рецепту, в полученном `DataFrame`.

```python
reviews_count = reviews.groupby("recipe_id").size().reset_index(name='count')

merged_outer = pd.merge(
    recipes[["id", "name"]],
    reviews_count,
    left_on="id",
    right_on="recipe_id",
    how="left"
)
merged_outer["count"] = merged_outer["count"].fillna(0).astype(int)
result_5_2 = merged_outer[["id", "name", "count"]].rename(columns={"id": "recipe_id"})
print(result_5_2.head(10))

result_5_2[result_5_2['count'] == 0].head()
result_5_2
```

5.3. Выясните, рецепты, добавленные в каком году, имеют наименьший средний рейтинг?

```python
ratings_with_year = reviews.merge(recipes[['id', 'year']], left_on='recipe_id', right_on='id')
year_avg_rating = ratings_with_year.groupby('year')['rating'].mean().reset_index(name='avg_rating')
min_rating_year = year_avg_rating.loc[year_avg_rating['avg_rating'].idxmin()]
print('Year with lowest average rating')
print(min_rating_year)
```
