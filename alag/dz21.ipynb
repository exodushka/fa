```python
#a 
eq1 = ((5 + I) * (7 -6*I)) / (3 + I)
eq1_alg = simplify(eq1)
eq1_mod = Abs(eq1)
eq1_arg = arg(eq1)
print(f"alg: {eq1_alg}")
print(f"mod: {eq1_mod} arg: {eq1_arg}")
print(f"trig: {eq1_mod}cos({eq1_arg}) + i*sin({eq1_arg})")
print(f"показательная: {eq1_mod}*exp(i*{eq1_arg})")
```

```python
import sympy as sp
import matplotlib.pyplot as plt
import cmath

I = sp.I

roots_a = [sp.exp(2*sp.pi*I*k/8) for k in range(8)]

z_b = -18 / (1 + I*sp.sqrt(3))
z_b_simplified = sp.simplify(z_b)
roots_b = [sp.root(z_b_simplified, 4, k) for k in range(4)]

expr_c = (1 - 5*I)/(1 + I) - 5*(1 + 2*I)/(2 - I) + 11
expr_c_simplified = sp.simplify(expr_c)
roots_c = [sp.root(expr_c_simplified, 3, k) for k in range(3)]

z_d = 1 - I
roots_d = [sp.root(z_d, 9, k) for k in range(9)]

fig, axes = plt.subplots(2, 2, figsize=(10, 10))

ax = axes[0, 0]
ax.axhline(y=0, color='k', linewidth=0.5)
ax.axvline(x=0, color='k', linewidth=0.5)
vals_a = [complex(r.evalf()) for r in roots_a]
xs_a = [v.real for v in vals_a] + [vals_a[0].real]
ys_a = [v.imag for v in vals_a] + [vals_a[0].imag]
ax.plot(xs_a, ys_a, 'r-', linewidth=1.5, alpha=0.7)
for v in vals_a:
    ax.plot(v.real, v.imag, 'ro', markersize=8)
    ax.plot([0, v.real], [0, v.imag], 'r-', alpha=0.3)
ax.grid(True, alpha=0.3)
ax.set_aspect('equal')

ax = axes[0, 1]
ax.axhline(y=0, color='k', linewidth=0.5)
ax.axvline(x=0, color='k', linewidth=0.5)
vals_b = [complex(r.evalf()) for r in roots_b]
xs_b = [v.real for v in vals_b] + [vals_b[0].real]
ys_b = [v.imag for v in vals_b] + [vals_b[0].imag]
ax.plot(xs_b, ys_b, 'b-', linewidth=1.5, alpha=0.7)
for v in vals_b:
    ax.plot(v.real, v.imag, 'bo', markersize=8)
    ax.plot([0, v.real], [0, v.imag], 'b-', alpha=0.3)
ax.grid(True, alpha=0.3)
ax.set_aspect('equal')

ax = axes[1, 0]
ax.axhline(y=0, color='k', linewidth=0.5)
ax.axvline(x=0, color='k', linewidth=0.5)
vals_c = [complex(r.evalf()) for r in roots_c]
xs_c = [v.real for v in vals_c] + [vals_c[0].real]
ys_c = [v.imag for v in vals_c] + [vals_c[0].imag]
ax.plot(xs_c, ys_c, 'g-', linewidth=1.5, alpha=0.7)
for v in vals_c:
    ax.plot(v.real, v.imag, 'go', markersize=8)
    ax.plot([0, v.real], [0, v.imag], 'g-', alpha=0.3)
ax.grid(True, alpha=0.3)
ax.set_aspect('equal')

ax = axes[1, 1]
ax.axhline(y=0, color='k', linewidth=0.5)
ax.axvline(x=0, color='k', linewidth=0.5)
vals_d = [complex(r.evalf()) for r in roots_d]
xs_d = [v.real for v in vals_d] + [vals_d[0].real]
ys_d = [v.imag for v in vals_d] + [vals_d[0].imag]
ax.plot(xs_d, ys_d, 'm-', linewidth=1.5, alpha=0.7)
for v in vals_d:
    ax.plot(v.real, v.imag, 'mo', markersize=8)
    ax.plot([0, v.real], [0, v.imag], 'm-', alpha=0.3)
ax.grid(True, alpha=0.3)
ax.set_aspect('equal')

plt.tight_layout()
plt.show()
```
