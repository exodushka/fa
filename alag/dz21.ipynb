```python
#a 
eq1 = ((5 + I) * (7 -6*I)) / (3 + I)
eq1_alg = simplify(eq1)
eq1_mod = Abs(eq1)
eq1_arg = arg(eq1)
print(f"alg: {eq1_alg}")
print(f"mod: {eq1_mod} arg: {eq1_arg}")
print(f"trig: {eq1_mod}cos({eq1_arg}) + i*sin({eq1_arg})")
print(f"показательная: {eq1_mod}*exp(i*{eq1_arg})")
```

```python
#b 
eq2 = ((1+I)**5) / (1 - I)**3
eq2_alg = simplify(eq2)
eq2_mod = Abs(eq2)
eq2_arg = arg(eq2)
print(f"alg: {eq2_alg}")
print(f"mod: {eq2_mod} arg: {eq2_arg}")
print(f"trig: {eq2_mod}cos({eq2_arg}) + i*sin({eq2_arg})")
print(f"показательная: {eq2_mod}*exp(i*{eq2_arg})")
```

```python
#c
eq3_left = ((-1+I*sqrt(3))**15) / ((1 - I) ** 20)
eq3_right = ((-1 -I*sqrt(3))**15) / ((1 + I) ** 20)
eq3 = eq3_left + eq3_right 
eq3_alg = simplify(eq3)
eq3_mod = Abs(eq3)
eq3_arg = arg(eq3)
print(f"alg: {eq3_alg}")
print(f"mod: {eq3_mod} arg: {eq3_arg}")
print(f"trig: {eq3_mod}cos({eq3_arg}) + i*sin({eq3_arg})")
print(f"показательная: {eq3_mod}*exp(i*{eq3_arg})")
```

```python
#a
z = symbols("z")
eq1 = Eq(z**2 + (2*I - 7)*z + 13 - I, 0)
solution = solve(eq1, z)
solution
```

```python
#b 
z1, z2 = symbols("z1, z2")
eq2_1 = Eq(2*z1 - (2+I)*z2, -1*I)
eq2_2 = Eq((4-2*I)*z1 -5*z2, -1 -2*I)
solution = solve([eq2_1, eq2_2], [z1,z2])
solution
```
