```python
from sympy import *
```

```python
t = symbols('t')

e1_p = Matrix([1, 2, -1, -2])
e2_p = Matrix([2, 3, 0, -1])
e3_p = Matrix([1, 2, 1, 4])
e4_p = Matrix([1, 3, -1, 0])

C = Matrix.hstack(e1_p, e2_p, e3_p, e4_p)

print("Матрица перехода C (столбцы - векторы e'_i):")
C
```

```python
det_C = C.det()
det_C
```

```python
x_old = Matrix([7, 14, -1, 2])

x_new = C.inv() * x_old

x_new
```

```python
#2
x_sym = symbols('x')

basis_e_prime = [(x_sym - 7)**i for i in range(5)]

P4 = 1 + 2*x_sym - 3*x_sym**2 + 4*x_sym**3 - 9*x_sym**4

cols = []
for poly in basis_e_prime:
    expanded_poly =expand(poly)
    coeffs = [expanded_poly.coeff(x_sym, i) for i in range(5)]
    cols.append(Matrix(coeffs))

C_poly = Matrix.hstack(*cols)

print("Матрица перехода от базиса e' к e:")
C_poly
```

```python

coords_P4_old = Matrix([P4.coeff(x_sym, i) for i in range(5)])

coords_P4_new = C_poly.inv() * coords_P4_old

coords_P4_new
```
