```python
import numpy as np
from sympy import *
init_printing()
def rrstr(x,n): # округление до n знаков после запятой
    fmt = '{:.'+str(n)+'f}'
    return fmt.format(x).replace('.',',')
```

1. На двумерной координатной плоскости заданы точки: 
$A(21,2)$, $B(2,26)$ и $C(48,48)$.
Пусть $D(X,Y)$ – ближайшая к $A$ точка прямой $BC$.
Найдите точку $D$ и расстояние $R$ от $A$ до $D$.
Укажите в ответе:
1) X, первую координату точки D;
2) Y, вторую координату точки D;
3) R, расстояние от A до D.<br/>

```python
A = Point(21, 2)
B = Point(2, 26)
C = Point(48, 48)
D = Line(B,C).projection(A)
```

```python
rrstr(D.x.evalf(),2)
```

```python
rrstr(D.y.evalf(),2)
```

```python
rrstr(float(A.distance(D)), 2)
```

2. Дана матрица: 
$A = \pmatrix{23 & 13 & 26\\ 29 & 22 & 27 \\ 10 & 25 & 24}$.
Пусть $Z$ – собственное значение матрицы $A$ с наибольшей мнимой частью.
Найдите $Z$ и соответствующий этому собственному значению собственный вектор $X$, 
первая координата которого равна 1, $X=(1,U,V)$.
В ответе укажите:
1) мнимую часть $Z$;
2) действительную часть $U$.<br/>

```python
A = np.array([
    [23, 13, 26],
    [29, 22, 27],
    [10, 25, 24]
])
vals, vecs = np.linalg.eig(A)
i = np.argmax(vals.imag)
Q = vals[i]
X = vecs[:, i]
X = X / X[0]
```

```python
rrstr(Q.imag, 3)
```

```python
rrstr(Q.imag, 3)
```
