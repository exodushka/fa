```python
#1
f1 = 2*x**4 - 3*x**3 + 4*x**2 - 5*x + 6
g1 = x**2 - 3*x + 1

h1, r1 = div(f1, g1)

check1 = expand(h1 * g1 + r1)
is_correct_1 = simplify(f1 - check1) == 0

print(f"Многочлен f(x): {f1}")
print(f"Многочлен g(x): {g1}")
print(f"Частное h(x): {h1}")
print(f"Остаток r(x): {r1}")
print(f"Проверка равенства (f == h*g + r): {is_correct_1}")
```

```python
#2
f2 = x**5 + 3*x**4 - 12*x**3 - 52*x**2 - 52*x - 12
g2 = x**4 + 3*x**3 - 6*x**2 - 22*x - 12

d = gcd(f2, g2)

are_coprime = d == 1

u, v, d_calc = gcdex(f2, g2)

bezout_check = simplify(f2 * u + g2 * v - d) == 0

print(f"{f2}")
print(f"{g2}")
print(f"{d}")
print(f"{are_coprime}")
print(f"{u}")
print(f"{v}")
print(f"{bezout_check}")
```

```python
#4a
f_4a = x**5 - a*x**2 - a*x + 1
x0 = -1
f_4a_at_x0 = f_4a.subs(x, x0)
df_4a = diff(f_4a, x)
df_4a_at_x0 = df_4a.subs(x, x0)


solution_4a = solve([f_4a_at_x0, df_4a_at_x0], a)
print(solution_4a)

a_val = solution_4a[a] if isinstance(solution_4a, dict) else solution_4a[0]
f_checked = f_4a.subs(a, a_val)
print(f"  f(x) = {f_checked}")
print(f"  f({x0}) = {f_checked.subs(x, x0)}")
print(f"  f'({x0}) = {diff(f_checked, x).subs(x, x0)}")
```
