```python
from sympy import *
import numpy as np
x = symbols("x")
```

```python
#1
f1 = 2*x**4 - 3*x**3 + 4*x**2 - 5*x + 6
g1 = x**2 - 3*x + 1

h1, r1 = div(f1, g1)

check1 = expand(h1 * g1 + r1)
is_correct_1 = simplify(f1 - check1) == 0

print(f"Многочлен f(x): {f1}")
print(f"Многочлен g(x): {g1}")
print(f"Частное h(x): {h1}")
print(f"Остаток r(x): {r1}")
print(f"Проверка равенства (f == h*g + r): {is_correct_1}")
```

```python
#2
f2 = x**5 + 3*x**4 - 12*x**3 - 52*x**2 - 52*x - 12
g2 = x**4 + 3*x**3 - 6*x**2 - 22*x - 12

d = gcd(f2, g2)

are_coprime = d == 1

u, v, d_calc = gcdex(f2, g2)

bezout_check = simplify(f2 * u + g2 * v - d) == 0

print(f"Многочлен f(x): {f2}")
print(f"Многочлен g(x): {g2}")
print(f"НОД(f, g): {d}")
print(f"Многочлены взаимно простые: {are_coprime}")
print(f"Многочлен u(x): {u}")
print(f"Многочлен v(x): {v}")
print(f"Проверка тождества Безу (f*u + g*v == d): {bezout_check}")
```
