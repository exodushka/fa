```python
from sympy import *
init_printing()
```

```python
#1
a1 = Matrix([2, 1, 3, -1])
a2 = Matrix([7, 4, 3, -3])
a3 = Matrix([1, 1, -6, 0])
a4 = Matrix([5, 7, 7, 8])
vectors = [a1, a2, a3, a4]

ortho = []
for v in vectors:
    w = v.copy()
    for u in ortho:
        w = w - (u.dot(v) / u.dot(u)) * u
    
    if w.dot(w) > 0:
        ortho.append(w)

orthonormal_basis = [v / sqrt(v.dot(v)) for v in ortho]

for i, b in enumerate(orthonormal_basis, 1):
    print(f"b_{i} = {list(b)}")
```

```python
#6.2
A = Matrix([[2, 1, 1, 3],
            [3, 2, 2, 1],
            [1, 2, 2, -4]])

L_basis = A.nullspace()

B = Matrix.hstack(*L_basis)
x = Matrix([7, -4, -1, 2])

proj = B * (B.T * B).inv() * B.T * x
orth_comp = x - proj

print("проекция", proj)
orth_comp
```

```python
#6.3
n = Matrix([2, -2, 3, -2, 2])
x = Matrix([3, 3, -1, 1, -1])

dist = Abs(n.dot(x)) / sqrt(n.dot(n))
print(dist.simplify())
```

```python
#6.4
x = Matrix([2, 2, 1, 1])
a1 = Matrix([3, 4, -4, -1])
a2 = Matrix([0, 1, -1, 2])

B = Matrix.hstack(a1, a2)
# Проекция вектора x на L
proj = B * (B.T * B).inv() * B.T * x

# cos(θ) = (x·proj) / (|x|·|proj|)
cos_theta = x.dot(proj) / (sqrt(x.dot(x)) * sqrt(proj.dot(proj)))
theta = acos(cos_theta)

print(f"угол между x и L: {theta.evalf()} рад ≈ {theta*180/pi.evalf()} град")
```

```python
#6.5
A = Matrix([45, 48])
B = Matrix([1, 4])
C = Matrix([4, 40])

v = C - B 

t = (A - B).dot(v) / v.dot(v)

D = B + t * v  
R = sqrt((A - D).dot(A - D))  

print(f"1) X = {D[0]}")
print(f"2) Y = {D[1]}")
print(f"3) R = {R}")
```

```python
A = Matrix([[24, 28, 20],
            [15, 21, 12],
            [10, 25, 11]])

eigvals = A.eigenvals()

Z = max([ev for ev in eigvals.keys() if im(ev) != 0], key=lambda x: float(im(x)))

U, V = symbols('U V')
X = Matrix([1, U, V])

eq = A * X - Z * X
sol = solve(list(eq), (U, V))

print(f"Im(Z) = {im(Z).evalf()}")
print(f"Re(U) = {re(sol[U]).evalf()}")
```

```python
import numpy as np

A = np.array([[24, 28, 20],
              [15, 21, 12],
              [10, 25, 11]])

eigvals, eigvecs = np.linalg.eig(A)

idx = np.argmax(eigvals.imag)
Z = eigvals[idx]

X = eigvecs[:, idx]
X = X / X[0]  

print(f"Im(Z) = {Z.imag}")
print(f"Re(U) = {X[1].real}")
```

```python
from sympy import Point, Line

A = Point(41, 44)
B = Point(13, 6)
C = Point(3, 42)

D = Line(B, C).projection(A)
R = A.distance(D)

print(f"X = {D.x.evalf()}")
print(f"Y = {D.y.evalf()}")
print(f"R = {R.evalf()}")
```
