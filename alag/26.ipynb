```python
#1
a1 = Matrix([2, 1, 3, -1])
a2 = Matrix([7, 4, 3, -3])
a3 = Matrix([1, 1, -6, 0])
a4 = Matrix([5, 7, 7, 8])
vectors = [a1, a2, a3, a4]

ortho = []
for v in vectors:
    w = v.copy()
    for u in ortho:
        w = w - (u.dot(v) / u.dot(u)) * u
    
    if w.dot(w) > 0:  # пропускаем нулевые векторы (если система линейно зависима)
        ortho.append(w)

# Нормировка каждого вектора
orthonormal_basis = [v / sqrt(v.dot(v)) for v in ortho]

for i, b in enumerate(orthonormal_basis, 1):
    print(f"b_{i} = {list(b)}")
```

```python
#6.2
# Матрица системы, задающей L
A = Matrix([[2, 1, 1, 3],
            [3, 2, 2, 1],
            [1, 2, 2, -4]])

# Находим базис подпространства L (ядро матрицы)
L_basis = A.nullspace()
print("Базис L:", L_basis)

B = Matrix.hstack(*L_basis)
x = Matrix([7, -4, -1, 2])

# Формула проекции: proj = B*(B^T*B)^(-1)*B^T * x
proj = B * (B.T * B).inv() * B.T * x
orth_comp = x - proj

print("Проекция x на L:", proj)
print("Ортогональная составляющая:", orth_comp)
```

```python
#6.3
# Нормаль к гиперплоскости и точка
n = Matrix([2, -2, 3, -2, 2])
x = Matrix([3, 3, -1, 1, -1])

# Формула расстояния до гиперплоскости n*x = 0: |n*x| / ||n||
dist = Abs(n.dot(x)) / sqrt(n.dot(n))
print("Расстояние:", dist.simplify())
```

```python
#6.4
x = Matrix([2, 2, 1, 1])
a1 = Matrix([3, 4, -4, -1])
a2 = Matrix([0, 1, -1, 2])

B = Matrix.hstack(a1, a2)
# Проекция вектора x на L
proj = B * (B.T * B).inv() * B.T * x

# cos(θ) = (x·proj) / (|x|·|proj|)
cos_theta = x.dot(proj) / (sqrt(x.dot(x)) * sqrt(proj.dot(proj)))
theta = acos(cos_theta)

print(f"Угол между x и L: {theta.evalf()} рад ≈ {theta*180/pi.evalf()} град")
```

```python
#6.5
A = Matrix([45, 48])
B = Matrix([1, 4])
C = Matrix([4, 40])

v = C - B  # направляющий вектор прямой BC
# Параметр проекции точки A на прямую BC
t = (A - B).dot(v) / v.dot(v)

D = B + t * v  # координаты точки D
R = sqrt((A - D).dot(A - D))  # расстояние

print(f"1) X = {D[0]}")
print(f"2) Y = {D[1]}")
print(f"3) R = {R}")
```
