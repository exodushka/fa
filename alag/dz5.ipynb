```python
def f(A, B, C, D):
    V = Matrix([
        [A[0], A[1], A[2], 1],
        [B[0], B[1], B[2], 1],
        [C[0], C[1], C[2], 1],   
        [D[0], D[1], D[2], 1]     
    ])
    V = V.det() / 6
    return abs(V)
A,B,C,D = [-1,1,2], [0,3,3], [1,-1,2],[0,0,1]
f(A, B, C, D)
```

```python
def solve_problem():
    n = 80000
    a11, ann = 0.99, 1.01
    h = (ann - a11) / (n - 1)
    a22 = a11 + h
    det_block = a11 * a22 - 0.57 * 0.03
    log_product = 0.0
    trace_inv = 0.0
    for i in range(3, n + 1):
        a_ii = a11 + (i - 1) * h
        log_product += np.log(a_ii)
        trace_inv += 1.0 / a_ii
    product_diag = np.exp(log_product)
    det_A = det_block * product_diag
    trace_inv += (a11 + a22) / det_block
    return det_A, trace_inv
determinant, trace_inverse = solve_problem()
print(f"1) Определитель матрицы A: {determinant:.3f}")
print(f"2) След матрицы, обратной к A: {trace_inverse:.3f}")
```

```python

def solve_matrix_problem():
    n = 43000
    
    q = (1.07) ** (1/(n-1))
    
    d = (1.10 - 1) / (n-1)
    diag_B = np.zeros(n)
    for i in range(n):
        diag_B[i] = 1 / (q ** i)
    first_row_B = np.zeros(n)
    for j in range(1, n):
        a_1j = 1 + j * d  
        a_jj = q ** j      
        first_row_B[j] = -a_1j / a_jj

    min_element = min(first_row_B[1:])
    
    total_sum = 0
    
    total_sum += np.sum(diag_B)
    
    total_sum += np.sum(first_row_B[1:])
    
    
    return min_element, total_sum

min_elem, sum_elem = solve_matrix_problem()

print(f"1) –{min_elem:.3f}".replace('.', ','))
print(f"2) –{abs(sum_elem):.1f}".replace('.', ','))
```
