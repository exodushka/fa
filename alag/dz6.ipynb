```python
from sympy import *
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
```

![](pic/6/6.1.png)

```python
n = 100

m = np.full((n,n), np.nan)

for i in range(n):
    for j in range(i + 1):
        if j == 0 or j == i:
            m[i][j] = 1
        else:
            m[i][j] = (m[i-1][j-1] + m[i-1][j]) % 2
plt.figure(figsize=(10, 10))
plt.imshow(m, interpolation='nearest', cmap=ListedColormap(["yellow", "blue"]))
plt.show()
```
