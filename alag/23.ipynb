```python
from sympy import *
from sympy.polys.polyfuncs import symmetrize

init_printing()

x1, x2, x3, x4 = symbols('x1 x2 x3 x4')
x, y, b, a, lam = symbols('x y b a lam')
c1, c2, c3, c4 = symbols('c1 c2 c3 c4')
s1, s2, s3, s4 = symbols('s1 s2 s3 s4')
```

```python
fa = x2*x1**3 + x3*x2**3*x1 + x3**3*x1 + x2*x3**3
sym_fa, rem_fa, _ = symmetrize(fa, [x1, x2, x3], formal=True)
sym_fa
check_fa = sym_fa.subs({s1: x1+x2+x3, s2: x1*x2+x1*x3+x2*x3, s3: x1*x2*x3})
check_fa
```

```python
fb = (3*x1 - x2 - x3)*(3*x2 - x1 - x3)*(3*x3 - x1 - x2)
sym_fb, _, _ = symmetrize(fb, [x1, x2, x3], formal=True)
sym_fb
check_fb = sym_fb.subs({s1: x1+x2+x3, s2: x1*x2+x1*x3+x2*x3, s3: x1*x2*x3})
check_fb - sym_fb
```
