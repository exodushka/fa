```python
def count_perm(l):
    inv = 0
    for i in range(len(l)):
        for j in range(i + 1, len(l)):
            if l[i] > l[j]:
                inv += 1
    return inv

a_p = [3, 5, 2, 1, 6, 4, 8, 7]
print(count_perm(a_p))
```
