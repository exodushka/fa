```python
k = 1
lambda_val = exp(2 * pi * I * k / 25)

A = Matrix([[4,5], [8,6]])
I_matrix = eye(2)

matrix = A + lambda_val * I_matrix

det = matrix.det()

real_part = re(det)
imag_part = im(det)

round(float(real_part), 3), round(float(imag_part), 3)
```
