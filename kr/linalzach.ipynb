```python
from sympy import *
init_printing()
x = symbols("x")
```

```python
k = 1
lambda_val = exp(2 * pi * I * k / 25)

A = Matrix([[4,5], [8,6]])
I_matrix = eye(2)

matrix = A + lambda_val * I_matrix

det = matrix.det()

real_part = re(det)
imag_part = im(det)

round(float(real_part), 3), round(float(imag_part), 3)
```

```python
p = x**12 +2*x**10 -2*x**8 +x**7 -x**6-2*x**4-x**2+2*x+2
roots = nroots(p)

sum_moduli = sum(abs(r) for r in roots) 

max_real_part = max(re(r) for r in roots)
round(float(sum_moduli), 3), round(float(max_real_part), 3)
```

```python
A = Matrix([
    [21, 23, 22],
    [11,17,15],
    [18,20,25]
])

eigenvalues = A.eigenvals()
Z = max(eigenvalues.keys(), key=lambda k: im(k))
eigenvectors_data = A.eigenvects()


for val, mult, vectors in eigenvectors_data:
    if val == Z:
        target_vector = vectors[0]
        break

U = target_vector[1] / target_vector[0]

imaa = float(im(Z))
reaa = float(re(U))
round(imaa, 3), round(reaa, 3)
```

```python
A = Point(48, 45)
B = Point(37, 6)
C = Point(22, 46)

line_BC = Line(B, C)
D = line_BC.projection(A)
R = A.distance(D)

round(float(D.x),2), round(float(D.y),2), round(float(R),2)
```
